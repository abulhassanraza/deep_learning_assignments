```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
x =  df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y , test_size= 0.30, random_state= 50)
```

```python
optimizer= 'rmsprop'
loss= 'binary_crossentropy'
metrics= ['accuracy']
```

```python
def model_build():
  model = Sequential()
  model.add(Dense(10, activation='tanh', input_shape=(30,)))
  model.add(Dense(8, activation= 'relu'))
  model.add(Dense(6, activation= 'relu'))
  model.add(Dense(1))
  model.compile(optimizer= optimizer, loss= loss, metrics=metrics)
  return model
```

```python
network = model_build()
```

```python
network.fit(x_train, y_train, epochs=100, batch_size= 512, validation_split= 0.20)
```

```python
np.random.seed(7)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=model_build, epochs=50, batch_size=512, verbose=1)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=2, shuffle=True, random_state=7)
results = cross_val_score(pipeline, x_train, y_train, cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
